```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense
import matplotlib.pyplot as plt
import os
import numpy as np
import seaborn as sns
from keras import layers, models, regularizers, optimizers
```

```python
# Set the paths to train, test, and validation folders
train_folder = 'Data/train'
test_folder = 'Data/test'
val_folder = 'Data/val'
```

```python
# Use ImageDataGenerator for loading and Rescale pixel values between 0 and 1
datagen = ImageDataGenerator(rescale=1./255)  
```

```python
# Downsampling the images to 128x128 pixels
downsampled_size = (128, 128)

# Set the batch size
batch_size = 32

# Load the train dataset with downsampling
train_generator = datagen.flow_from_directory(
    train_folder,
    target_size=downsampled_size,
    batch_size=5216,
    class_mode='binary',
    shuffle=True
)

# Load the test dataset with downsampling
test_generator = datagen.flow_from_directory(
    test_folder,
    target_size=downsampled_size,
    batch_size=624,
    class_mode='binary',
    shuffle=False
)

# Load the validation dataset with downsampling
val_generator = datagen.flow_from_directory(
    val_folder,
    target_size=downsampled_size,
    batch_size=16,
    class_mode='binary',
    shuffle=False
)
```

```python
def plot_class_counts(labels, title):
    # Get the counts for each class
    class_labels, counts = np.unique(labels, return_counts=True)
    # Calculate the percentage of value counts
    total_count = np.sum(counts)
    percentages = counts / total_count * 100
    # Plot the results
    plt.figure(figsize=(12,10))
    plt.title(f'{title}', fontsize='16')
    plt.xlabel("Classes")
    plt.ylabel("Count")
    # Replace "1" with "Normal" in class_labels list
    class_labels = ['Pneumonia', 'Normal']
    # Add value counts as text on top of each bar
    for i in range(len(class_labels)):
        plt.text(i, counts[i], f"{counts[i]}\n{percentages[i]:.2f}%", ha='center', va='bottom')
    sns.barplot(x=class_labels, y=counts)
    plt.show()
```

```python
labels = train_generator.classes
plot_class_counts(labels, 'Training Data Distribution')
```

```python
test_labels = test_generator.classes
plot_class_counts(test_labels, 'Test Data Distribution')
```

```python
vals_labels = val_generator.classes
plot_class_counts(vals_labels, 'Validation Data Distribution')
```

## Modelling

### Baseline Model - A densely connected network

```python
# Build a baseline fully connected model
np.random.seed(123)
model_1 = models.Sequential()
model_1.add(layers.Dense(20, activation='relu', input_shape=(49152,))) # 1 hidden layers
model_1.add(layers.Dense(7, activation='relu'))
model_1.add(layers.Dense(1, activation='sigmoid'))
```

```python
model_1.compile(optimizer='sgd',
              loss='binary_crossentropy',
              metrics=['accuracy'])
```

```python
model_1.summary()
```

```python
results_1 = model_1.fit(train_img,
                    train_labels,
                    epochs=25,
                    batch_size=20,
                    validation_data=(val_img, val_labels))
```

Now, let's inspect the model's performance and see if we detect any overfitting or other issues. In the cell below, create two plots:

* The `loss` and `val_loss` over the training epochs
* The `acc` and `val_acc` over the training epochs

```python
def visualize_training_results(results):
    history = results.history
    plt.figure(figsize=((10,8)))
    plt.plot(history['val_loss'])
    plt.plot(history['loss'])
    plt.legend(['val_loss', 'loss'])
    plt.title('Loss')
    plt.xlabel('Epochs')
    plt.ylabel('Loss')
    plt.show()
    
    plt.figure(figsize=((10,8)))
    plt.plot(history['val_accuracy'])
    plt.plot(history['accuracy'])
    plt.legend(['val_acc', 'acc'])
    plt.title('Accuracy')
    plt.xlabel('Epochs')
    plt.ylabel('Accuracy')
    plt.show()
```

```python
visualize_training_results(results_1)
```

### Model 3 - CNN with Dropout & L2 Regularization

```python

model_3 = models.Sequential()
model_3.add(layers.Conv2D(32, (3, 3), activation='relu', input_shape=(128, 128, 3)))
model_3.add(layers.MaxPooling2D((2, 2)))
model_3.add(layers.Dropout(0.25))  # Dropout layer with a rate of 0.25

model_3.add(layers.Conv2D(64, (3, 3), activation='relu', kernel_regularizer=regularizers.l2(0.01)))  # L2 regularization with a factor of 0.01
model_3.add(layers.MaxPooling2D((2, 2)))
model_3.add(layers.Dropout(0.25))

model_3.add(layers.Conv2D(64, (3, 3), activation='relu', kernel_regularizer=regularizers.l2(0.01)))
model_3.add(layers.MaxPooling2D((2, 2)))
model_3.add(layers.Dropout(0.25))

model_3.add(layers.Flatten())
model_3.add(layers.Dense(64, activation='relu', kernel_regularizer=regularizers.l2(0.01)))
model_3.add(layers.Dense(1, activation='sigmoid'))
```

```python
model_3.compile(loss='binary_crossentropy',
              optimizer='adam',
              metrics=['accuracy'])
```

```python
results_3 = model_3.fit(train_images,
                    train_labels,
                    epochs=20,
                    batch_size=32,
                    validation_data=(val_images, val_labels))
```

```python
visualize_training_results_cnn(results_3)
```
